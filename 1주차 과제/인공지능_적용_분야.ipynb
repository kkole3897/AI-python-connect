# 인공지능 적용분야 

## 0. 개요

현재 인공지능은 다양한 분야에서 사용되고, 연구되고 있습니다. 그 중 언어, 음성, 이미지, 자율주행 분야에서 인공지능 기술이 적용된 제품이나
서비스를 찾아보고 정리하고자 합니다.

## 1. 언어: 구글 번역

기계 번역이 언어 분야에서의 인공지능의 가장 기본적인 Task라고 생각합니다. 기계 번역을 위해 구글이 개발한 트랜스포머 기반 BERT 모델은 기계 번역에서 새로운 패러다음을 제시했고, 이후 나오는 새로운 언어 분야에서의 모델들은 거의 BERT의 변형입니다.

### 1.1. 언어 분야에 대한 개인적인 목표

제가 언어 분야에서 가장 관심이 있는 것은 QA입니다. 현재 검색 서비스는 사용자가 입력한 쿼리와 유사도가 가장 높은 문서를 분석해서 해당 문서를 보여줍니다. 하지만 이는 사용자의 질문에 대한 답을 직접해주지는 않습니다. 예전부터 사용자의 질문에 대한 답을 직접적으로 해줄 수 있는 검색 서비스에 대해 관심을 갖고 있었고, 기회가 된다면 직접 개발해보고 싶었습니다. 이와 관련된 분야가 QA라는 것을 알게 되었고, 언어 지능 분야에서 주된 연구 분야라고 알고 있습니다. 한국어의 경우는 전세계적으로 가장 널리 연구되고 있는 영어랑 언어 체계가 상이해서 한국어 자체의 새로운 모델을 개발해야하는 경우가 많습니다. 최근 LG CNS에서 공개한 QA 연구를 위해 공개된 언어 데이터셋 'KorQuad'를 이용해서 직접 한국어 QA 모델을 개발하고 싶습니다.

## 2. 음성: 알렉사, 클로바 AI 스피커

음성 분야의 인공지능하면 가장 먼저 떠오르는 것은 인공지능 스피커입니다. 스마트폰이 터치 기반의 인터페이스의 혁신을 가져다 주었다면 인공지능 스피커는 음성이라는 새로운 인터페이스로 인간이 기기를 조작할 수 있도록 혁신하고 있습니다. 인공지능 스피커에 적용되는 음성 관련 인공지능을 보자면 우선 음성 인식이 있습니다. 사람이 하는 말을 정확하게 인식해야 각종 기기들이 해당 명령에 대한 정확한 동작을 할 수 있기 때문에 가장 기본이 되는 기술이라 할 수 있습니다. 그리고 최근에는 음성 합성 기술이 중요한 화두로 떠오르고 있습니다. 각종 전자제품들이 단순히 명령에 대한 응답으로 작동만 하는 것이 아니라 인간과 소통하도록 발전하고 있습니다. 자연스러운 음성을 만드는게 음성 합성의 목표이기 때문에 기기와 인간의 상호작용에 점점 더 중요한 기술로 자리잡고 있습니다. 또한 음성 합성 기술은 인터넷 개인 방송에서 채팅을 읽어주는 방식으로 사용자에게 또다른 재미를 가져다 주고 있습니다.

### 2.1 음성 분야 개인적인 목표

아직 음성 분야에 대한 이해는 거의 없는 상태이지만 인공지능 작곡/편곡에 대해 흥미를 갖고 있습니다. 음악을 듣던 중 하나의 노래라도 여러 방식으로 실시간으로 편곡되어 스트리밍하는 서비스가 있다면 음악 감상을 좀 더 재밌게 할 수 있을 거라고 생각해왔습니다. 딥러닝 공부를 하면서 이미지에 대해 Style Transfer가 가능하다면 음악에도 이를 적용할 수 있지 않을까라는 호기심이 생겼습니다. 아직 시도해보지 않았지만 음성 쪽으로 공부하게 되면 한번쯤 시도해볼만한 가치가 있다고 생각합니다.

## 3. 이미지

개인적으로 가장 관심 있는 분야입니다. 이미지에서 가장 기본적인 Task는 객체 인식과 분류입니다. 2012년 이미지넷 대회에서 딥러닝을 사용한 모델이 압도적인 차이로 우승을 한 이후 컴퓨터 비전 분야에서 딥러닝은 굉장히 중요해졌습니다.

### 3.1. 얼굴 인식

아마 이미지 분야에서의 인공지능에서 인간이 가장 먼저 정복한 분야가 아닌가 싶습니다. 이제 더이상 Challenge한 영역은 아니지만 이미지 분야에서 초기부터 굉장히 중요한 분야였습니다. 최근에는 많은 스마트폰에서 얼굴 인식을 보안 절차로 채용하고 있는 경우가 많아, 많은 사람들이 직접 사용해본 기술 중에 하나라고 할 수 있습니다.

### 3.2. 사진 보정 어플

여러 사진 보정 기술 중에서도 사진의 스타일 자체를 바꿔주는 기술을 재밌게 봤습니다. 처음 관련 어플을 사용했을 땐 이게 인공지능이라는 생각은 전혀 해보지 못 했지만 이 또한 인공지능이 사용된 기술 중 하나라고 알게 되었습니다. 그리고 최근에는 GAN을 이용해서 단순히 색이나 스타일을 변형하는 것 뿐만 아니라 완전히 새로운 이미지를 만드는 것을 보면서 사람들이 인공지능이라고 인식하지 못하는 사이에 재미를 줄 수 있는 분야라고 생각합니다.

### 3.3. 이미지 분야 개인적인 목표

작년 CVPR에 올라온 논문 중 하나를 읽고 직접 저자의 깃헙에 올라온 코드를 이용해서 학습시키고 있는 모델이 있습니다. 이 모델을 이용해서 상품 진열대를 인식해 재고 관리를 하던, 고객들이 원하는 상품을 오프라인에서 더욱 쉽게 찾을 수 있도록 하는 어플리케이션을 직접 만들어보고 싶습니다.

## 4. 자율주행

자율주행은 여러 인공지능이 적용된 분야를 종합한 분야라고 볼 수 있습니다. 완벽한 자율주행을 위해선 인식, 판단, 제어가 반드시 필요합니다.

인식에서 가장 중요한 분야는 이미지 분야라고 생각합니다. 인식은 카메라를 통해 주변 상황을 정확하게 인식하는 것인데 이는 딥러닝을 통해 학습된 모델을 이용합니다. 이미 이미지 인식은 컴퓨터가 사람의 능력을 뛰어넘었기 때문에 큰 문제가 되지 않는다고 생각하지만 판단과 제어는 여전히 해결이 필요한 부분이라고 생각합니다. 판단과 제어를 위해서는 직접 사람의 운전 데이터를 이용해서 모델을 학습시켜야 하는데 아직 여기에 한계점이 많이 있는 것으로 보입니다.
